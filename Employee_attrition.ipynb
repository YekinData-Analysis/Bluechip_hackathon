```python
#TURNING FOR BEST PARAMETER VALUE

from sklearn.preprocessing import StandardScaler, LabelEncoder

X = dframe.drop(columns=['Attrition_Score'])
y = dframe['Attrition_Score']

scaler = StandardScaler()
X = scaler.fit_transform(X)

classifiers = {
    'Random Forest': (RandomForestClassifier(),
                      {'n_estimators': [100, 200], 'max_depth': [10, 20]}),
    'Extra Trees': (ExtraTreesClassifier(),
                    {'n_estimators': [100, 200], 'max_depth': [10, 20]}),
    'Gradient Boosting': (GradientBoostingClassifier(),
                          {'n_estimators': [50, 100], 'learning_rate': [0.01, 0.2]}),
    'Hist Gradient Boosting': (HistGradientBoostingClassifier(),
                               {'max_iter': [100, 200], 'learning_rate': [0.01, 0.2]}),
    'XGBoost': (XGBClassifier(),
               {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.001,0.0001],'max_depth':[100,200,500]}),
    'Logistic Regression': (LogisticRegression(solver='liblinear'),
                           {'C': [0.1, 1, 10]}),
    'DecisionTreeClassifier':(DecisionTreeClassifier(),{'max_depth':[10,200],'criterion':["gini", "entropy"]})
}
```

```python
#FITTING IN

n_splits = 5
stratified_kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

for clf_name, (clf, param_grid) in classifiers.items():
    grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, scoring='roc_auc', cv=stratified_kf)
    grid_search.fit(X, y)

    best_params = grid_search.best_params_
    best_auc = grid_search.best_score_
    print(f"{clf_name} - Best Hyperparameters: {best_params}, Best AUC: {best_auc:.4f}")
```
